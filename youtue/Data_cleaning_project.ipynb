# 11. Contract

### player's contract details

```python
#show the type of the column
fifa['Contract'].dtype
```

```python
#show the unique values if that column
fifa['Contract'].unique()
```

we want to make 3 columns that represent the start date and  the end date and the contract length

```python
#we will show the the rows that has the string on laon or the string free to see if all of them arr in the same format
for index, row in fifa.iterrows():
    if row['Contract']=='Free' or 'On Loan' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract=='Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date  = contract.split(' ~ ')
        contract_length = int(end_date) - int(start_date)
    return start_date, end_date, contract_length   

#apply the fun on Contract column and create new columns

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x : pd.Series(extract_contract_info(x)))

#loop for inserting the three columns in the fifa dataframe
for i in range(len(new_cols)):
    fifa.insert(loc = fifa.columns.get_loc('Contract')+1+i , column = new_cols[i], value = new_data[i])
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
#create a function that categorize the Contract cloumn
def categorize_contract_status(contract):
    if contract=='Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

#Add contract status column
fifa.insert(loc= fifa.columns.get_loc('Contract Length(years)')+1, column = 'Contract_Status', value = fifa['Contract'].apply(categorize_contract_status))
```

# 13. Height

### player's height in centimetres

```python
#show the data type of the column
fifa['Height'].dtype
```

```python
#show the unique values
fifa['Height'].unique()
```

As you can see there are height in cm and others in feet/inche

### 1 foot = 30.48 cm
### 1 inche = 2.54 cm
### 1 foot = 12 inches so to convert from feet to inches multiply by 12

```python
#create a function that convert all the height into centimeters
def convert_height(height):
    if 'cm' in height:
        return int(height.strip('cm'))
    else:
        feet, inches = height.split("\'")
        total_incehs = int(feet) * 12 + int(inches.strip('"'))
        return round(total_incehs * 2.54)

#apply the function to the Height column    
fifa['Height'] = fifa['Height'].apply(convert_height)
#recheck the unique values
fifa['Height'].unique()
```

```python
#let's rename the height column name to declare it's in cm 
fifa.rename(columns={"Height" : "Hieght(cm)"}, inplace=True)
fifa.sample(1)
```

```python
#recheck the type of the height column
fifa['Hieght(cm)'].dtype
```

the datatype of the Height column is now suitable

# 14. Weight


### player's weight in kilograms

```python
#show the data type of the column
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

As we can see that there are weight in kilograms and others in pound(lbs)
so let's correct this

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs'))
        return round(pounds/2.205)

#apply the function to the Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
#recheck the unique values
fifa['Weight'].unique()
```

All the weights have been converted into kg

```python
#let's rename the weight column name to declare it's in kg 
fifa.rename(columns={'Weight':'Weight(kg)'}, inplace=True)
fifa.sample(1)
```

```python
#recheck the type of the weight column
fifa['Weight(kg)'].dtype
```

the datatype of the Weight column is now suitable
